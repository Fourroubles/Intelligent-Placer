```python
from IntelegentPlacer import preprocessingImage as pi
import matplotlib.pyplot as plt
from skimage.color import rgb2gray, rgba2rgb
import cv2 as cv
from PIL import Image
```

```python
path = 'images'
image = pi.get_all_images(path)
```

Преобразуем изображение в чб

```python
image_grey = pi.convert_to_BW(image[10])
```

```python
image_threshold = pi.convert_to_trashold(image_grey, 150)
```

Находит контуры изображения

```python
counter = pi.find_contours(image_threshold)
```

```python
# рисуем контуры на исходном изображении
image_copy = image[10].copy()
cv.drawContours(image=image_copy, contours=counter, contourIdx=-1, color=(255, 255, 100), thickness=2, lineType=cv.LINE_AA)

# визуализировать двоичное изображение
plt.imshow(image_copy)
```
